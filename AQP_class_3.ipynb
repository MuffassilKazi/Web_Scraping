```python
import pandas as pd
import os
from bs4 import BeautifulSoup as bs
from IPython.core.display import display,html
import requests
import sys
import csv
```

```python
def met_data (month, year):
    
    file_html=open(f"data/html/{year}/{month}.html",'r')
    plain_text=file_html.read()
    
    temp_d=[]
    
    soup=bs(plain_text)
    
    for table in soup.find_all('table',attrs={'class':"medias mensuales numspan"}):
        for tbody in table:
            for tr in tbody:
                a=tr.get_text()
                temp_d.append(a)
#         return temp_d
    
       
    final_d=[]
    row=len(temp_d)/15
    
    for i in range(round(row)):
        temp=[]
        for j in range(0,15):
            temp.append(temp_d[0])
            temp_d.pop(0)
        final_d.append(temp)
        
    final_d.pop(0)
    final_d.pop(len(final_d)-1)
        
    for i in range(len(final_d)):
        final_d[i].pop(14)
        final_d[i].pop(13)
        final_d[i].pop(12)
        final_d[i].pop(11)
        final_d[i].pop(10)
        final_d[i].pop(6)
        final_d[i].pop(0)
        
        
    return final_d
    
    
            
```

```python
if not os.path.exists('data/Real_data'):
    os.makedirs('data/Real_data')
    
    
for year in range(2013, 2018+1):
    final_data=[]
    with open (f'data/Real_data/real_{year}.csv','w') as f:
        data=csv.writer(f,dialect='excel')
        data.writerow(
        ['T','TM','Tm','SLP','H', 'VV','V', 'VM', 'PM 2.5']
        )
        
    for month in range(1,13):
        temp= met_data(month,year)
        final_data+=temp
    
    
    pm=getattr(sys.modules[__name__],f'avg_data_{year}')()
#     print((pm))

    
#     print(final_data[1])
    for i in range(len(final_data)-1):
        final_data[i].insert(8,pm[i])
        
#     print(final_data[1])
    
    with open (f'data/Real_data/real_{year}.csv','a') as f:
        data=csv.writer(f,dialect='excel')
        for row in final_data:
            flag=0
            for elem in row:
                if elem=="-" or elem=="":
                    flag=1
            if flag != 1:
                data.writerow(row)
        
```

```python
def year_range(start, end):
    df_list=[]
    for year in range(start, end+1):
        df=pd.read_csv(f'data/Real_data/real_{year}.csv')
        df_list.append(df)
    return pd.concat(df_list, ignore_index=True)
```

```python
final_csv=year_range(2013,2015)
```

```python
all_final_csv=year_range(2013,2018)
```
